## Step 2: Import Necessary Libraries



```python
import pandas as pd
from pycaret.clustering import setup, create_model, assign_model, plot_model
```

## Step 3: Load Your Dataset



```python
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Display the first few rows of the dataset
df.head()
```

## Step 4: Set Up the PyCaret Environment



```python
# Initialize the clustering setup with preprocess=False
cluster_setup = setup(data=df, normalize=True, session_id=123, preprocess=False)
```

## Step 5: Create a DBSCAN Model



```python
# Create a DBSCAN model
dbscan_model = create_model('dbscan')
```

## Step 6: Assign Clusters to the Dataset



```python
# Assign clusters to the dataset
clustered_data = assign_model(dbscan_model)

# Display the dataset with cluster assignments
clustered_data.head()
```

## Step 7: Visualize the Clusters

```python

# Plot the clusters (Supported for DBSCAN)
plot_model(dbscan_model, plot='cluster')

# Distribution of clusters (Also supported for DBSCAN)
plot_model(dbscan_model, plot='distribution')

# Silhouette plot (Unsupported for DBSCAN; REMOVE THIS LINE)
# plot_model(dbscan_model, plot='silhouette')



# Save the clustered dataset to a CSV file
clustered_data.to_csv("clustered_data.csv", index=False)
print("Clustered data saved as clustered_data.csv")
```
